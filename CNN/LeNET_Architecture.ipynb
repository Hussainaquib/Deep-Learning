# **LeNet Architecture**

```python
from keras.models import Sequential
from keras.layers import Conv2D, AveragePooling2D, Flatten, Dense

model = Sequential()

# First Convolutional Layer followed by Average Pooling
model.add(Conv2D(6, kernel_size=(5,5), padding='valid', activation='tanh', input_shape=(32,32,1)))
model.add(AveragePooling2D(pool_size=(2, 2), strides=2, padding='valid'))

# Second Convolutional Layer followed by Average Pooling
model.add(Conv2D(16, kernel_size=(5,5), padding='valid', activation='tanh'))
model.add(AveragePooling2D(pool_size=(2, 2), strides=2, padding='valid'))

# Flattening the output from the convolutional layers
model.add(Flatten())

# Fully connected layers
model.add(Dense(120, activation='tanh'))
model.add(Dense(84, activation='tanh'))

# Output layer for classification (10 classes for MNIST)
model.add(Dense(10, activation='softmax'))
```

```python
model.summary()
```
