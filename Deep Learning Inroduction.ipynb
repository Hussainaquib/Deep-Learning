## What is Deep Learning ?
Several Deep Learning Definitions
- Deep Learning is a subfield of AI and ML that is inspired by the structure of human brain.
- Deep Learning algorithms attempts to draw similar conclusion as humans would by continually analysing data with a given logical structure called Neural Network.
- Deep Learning is part of a broader family of ML methods based on ANN with representative learning(Feature Learning), here deep learning algorithm automatically work on feature engineering.
- Deep Learning algorithms uses multiple layers to progressively extract higher-level features from the row input. for example, in image processing the lower layers may identify edges, while higher layers may identify the concepts relevant to a human such as digits or letters or faces.

- In ML the relation between the input and output is depends on statistical technique.
- In DL the relation between the input and output is depends on logical structure called Neural network and here the fundamental unit is called perceptron.

## Types of Neural Network
- Artificial Neural Network (ANN)
- Recurrent Neural Network (RNN)
    - work on speech and text data
- Convolutional Neural Network (CNN)
    - work on image data
- Generative Adversarial Network (GAN)
    - generate a text and image

## Deep Learning VS Machine Learning
- Data Dependency
    - for handling with large data we use Deep Learning, Because large data == better Model.
- Hardware Dependency
    - for sm0othly run the model/code we use GPU, but GPUs are very costly.
- Training Time
    - here the model is complex hence training time is very high, but prediction time is fast (In ML for KNN prediction time is high).
- Feature Selection
    - In ML feature is manually extract but in DL feature engineering is automatically done by DL algorithm.
- Interpretability
    - In DL interpretability is very low but in ML it is high.
